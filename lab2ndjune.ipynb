```python
r = requests.get('https://en.wikipedia.org/wiki/Python')
r.status_code
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
```

Retrieve an arbitrary Wikipedia page of "Python" and create a list of links on that page: url ='https://en.wikipedia.org/wiki/Python'

```python
soup.find('div', attrs={'class': 'mw-parser-output'}).find_all('a')
```

```python
df=soup.find('div', attrs={'class': 'mw-parser-output'}).find_all('a')
```

```python
titles = []
link = []

for i in df:
    titles.append(i.get_text())
    link.append(i.get('href'))
```

```python
link
```

```python
new_df = pd.DataFrame({'titles': titles, 'link': link})
```
